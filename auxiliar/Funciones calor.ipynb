```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, np.pi, 0.01)
relu = [0] + [0.3 if a < np.pi/2 else 0.8 for a in x] + [0]
x = np.append(x, np.pi)
x = np.append(x, np.pi+0.01)
```

```python
plt.plot(x, relu)
```

```python
x = np.arange(0, np.pi, 0.01)
relu = [0] + [0.3 if 0.2 < a < np.pi/2 else 0.8 if np.pi/2 < a < np.pi-0.2 else 0 for a in x] + [0]
x = np.append(x, np.pi)
x = np.append(x, np.pi+0.01)
```

```python
plt.plot(x, relu)
```

```python
x = np.arange(0, np.pi, 0.01)
relu = [0] + [2*a/np.pi if a < np.pi/2 else 2 - 2*a/np.pi for a in x] + [0]
x = np.append(x, np.pi)
x = np.append(x, np.pi+0.01)
```

```python
plt.plot(x, relu)
```
